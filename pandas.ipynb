```python
#Reshaping dataframes using pivot, stack

df = pd.DataFrame({"Region":['North','West','East','South','North','West','East','South'],
          "Team":['One','One','One','One','Two','Two','Two','Two'],
          "Revenue":[7500,5500,2750,6400,2300,3750,1900,575],
            "Cost":[5200,5100,4400,5300,1250,1300,2100,50]})
df
```

```python
df.pivot(index= 'Region', columns = 'Team', values = 'Revenue') #pivoting rows to columns
```

```python
df2 = df.set_index(['Region', 'Team'])
stacked = pd.DataFrame(df2.stack()) #opposite of pivot(pivoting column labels to rows)
stacked
```

```python
#Unstacking( pivoting row labels back to columns)
stacked.unstack('Region')
```

```python
#Melt allows you to reformat your dataframe to identify columns as "ID variables",
#while transforming all other columns, or "measure variables" to the row level.

df.melt(id_vars=['Region', 'Team'], var_name='value type')
```

```python
# Aggregation with pivot_table

df.pivot_table(index = 'Team', columns = 'Region', values = 'Revenue') #mean to aggregate(acts as a pandas groupby, then unstacks
#by the variable you specify in columns)
```

```python
df = pd.DataFrame({"Species":['Chinook','Chum','Coho','Steelhead','Bull Trout'],
             "Population":['Skokomish','Lower Skokomish','Skokomish','Skokomish','SF Skokomish'],
             "Count":[1208,2396,3220,6245,8216]})
df
```

```python
#Binning numerical data from pd.cut

bins = [0, 2000, 4000, 6000, 8000, np.inf]
labels = ['Low Return', 'Below Avg Return', 'Avg Return', 'Above Avg Return', 'High Return']
```

```python
df['count_category'] = pd.cut(df['Count'], bins, labels=labels)
df
```

```python
# Map species to Endangered Species

fed_status ={"Chinook":"Threatened",
"Chum":"Not Warranted",
"Coho":"Not Warranted",
"Steelhead":"Threatened"}
```

```python
df['Federal Status'] = df['Species'].map(fed_status) 
df   
```

```python
df['Count Category'] = pd.Categorical(df['count_category'], ordered = True, categories = labels)
df['Count Category']
```

```python
df.sort_values(by=['Count Category'], ascending = False)
```
